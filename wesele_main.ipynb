```python
cena_sniadania = 40
cena_pokoj = 180
```

```python
hotel["sniadania_dni"] = hotel["sniadania_dni"].fillna(0)
hotel["ilosc_pokoi"] = hotel["ilosc_pokoi"].fillna(0)
hotel["ilosc_dob"] = hotel["ilosc_dob"].fillna(0)
hotel["wartosc"] = hotel["wartosc"].fillna(0)
hotel["cena"] = hotel["ilosc_dob"] * cena_pokoj + hotel["sniadania_dni"] * hotel["liczba_osob"] * cena_sniadania
hotel
```

```python
hotel["wartosc"] = hotel["ilosc_pokoi"] * hotel["cena"]
hotel.at[7,"wartosc"] = hotel.at[7,"cena"]
hotel
```
